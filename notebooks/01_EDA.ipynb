```python
import sys
import os
sys.path.append(os.path.join(os.path.dirname(os.getcwd())))

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from app.data_loader import load_data
```

```python
# Charger les données
df = load_data()

# Aperçu du dataset
print("Dimensions :", df.shape)
display(df.head())
```

```python

# Vérification des types de colonnes
print("\nInfo dataset :")
print(df.info())

# Valeurs manquantes
print("\nValeurs manquantes par colonne :")
print(df.isna().sum())
```

```python

# Quelques stats descriptives
print("\nStatistiques descriptives :")
display(df.describe())
```

```python
# Exemple : Distribution des ventes
plt.figure(figsize=(8,5))
sns.histplot(df["Sales"], bins=50, kde=True)
plt.title("Distribution des ventes")
plt.xlim(0, 3000) # je limite l'axe des x pour une meilleure visualisation
plt.axvline(246.49, color='red', linestyle='--', label='Moyenne des ventes')
plt.legend()
plt.show()
```

```python
# Exemple : Ventes par région
plt.figure(figsize=(12,5))
sns.barplot(x="Region", y="Sales", data=df, estimator=sum)
plt.title("Ventes totales par région")
plt.show()
```
